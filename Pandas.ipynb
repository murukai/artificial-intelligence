## 1. By Default 
- matches the data at common column
- performs inner merge

```python
pd.merge(pers, marks_copy)
```

1. Inner join :include only the common primary keys   | left - present and right - present | 103,104,106,107,109,110,111,117,119
2. Outer join :includes all the primary keys          | left - present or right - present |
3. left  join : includes all primmmary keys from left | left - present 
4. right join : includes all primary keys from right  | right - present 

```python
pd.merge(pers, marks_copy, how = 'inner', on = 'Enrol No.')
```

```python
pd.merge(pers, marks_copy, how = 'outer')
```

```python
pd.merge(pers, marks_copy, how = 'left')
```

```python
pd.merge(pers, marks_copy, how = 'right')
```

```python
pers_short['Enrol No.'].values
```

```python
marks_copy['Enrol No.'].values
```

- inner-> 103,104, 110,111,117,119 -->  common ones 
- outer -> 102, 103,104,105,106,107,109,110,111,117,119 -> all ( including common)
- left -> pers_short['Enrol No.'].values
- right ->marks_copy['Enrol No.'].values

```python
def data_range(x):
    return x.max()-x.min()
```

```python
car.groupby('Type').agg({'Make' : 'nunique', 'MSRP' :['mean','std'],'Invoice':data_range}).round(2)
```
